```python
import pandas as pd

# 엑셀 파일 경로 설정
file_path = '데이터셋 (1).xlsx'

# 엑셀 파일 불러오기
excel_data = pd.ExcelFile(file_path)

# 첫 번째 시트 로드
df = excel_data.parse('시트1')

# '장점' 열 확인
advantages = df['장점'].dropna()
print(advantages.head())  # 데이터가 잘 불러와졌는지 확인
```

```python
from sklearn.feature_extraction.text import CountVectorizer
stopwords = ['및', '그리고', '에서', '있는', '하여', '하는', '입니다', '있습니다']  # 필요한 불용어를 추가

# CountVectorizer를 활용해 공백 기반 단어 빈도 계산
vectorizer = CountVectorizer(stop_words=stopwords, max_features=20)
X = vectorizer.fit_transform(advantages)

# 키워드와 빈도를 데이터프레임으로 정리
keywords = vectorizer.get_feature_names_out()
frequencies = X.toarray().sum(axis=0)

keywords_df = pd.DataFrame({'Keyword': keywords, 'Frequency': frequencies})
keywords_df = keywords_df.sort_values(by='Frequency', ascending=False).reset_index(drop=True)
print(keywords_df)
```

```python
from konlpy.tag import Okt
okt = Okt(jvmpath='C:/Program Files/Java/jdk-23/bin/server/jvm.dll')

positive_words = ['좋다', '훌륭하다', '즐겁다', '분위기', '좋은', '자유', '복지', '좋고', '최고', '식당', '점심']  # 긍정적인 단어 목록
negative_words = ['나쁘다', '힘들다', '어렵다']  # 부정적인 단어 목록

def sentiment_analysis(text):
    tokens = okt.morphs(text)
    score = 0
    for token in tokens:
        if token in positive_words:
            score += 1
        elif token in negative_words:
            score -= 1
    return score

df['Sentiment'] = df['장점'].apply(sentiment_analysis)
print(df[['장점', 'Sentiment']])
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# 샘플 데이터 (여기에 실제 데이터를 불러오세요)
# df = pd.read_excel('your_file.xlsx')
# 데이터는 '장점'과 '별점'라는 열이 있다고 가정
reviews = df[['장점', '별점']].dropna()

# 한국어 불용어 리스트 (필요에 따라 확장 가능)
korean_stop_words = ['및', '에서', '있다', '으로', '수', '있는', '하는', '그', '이', '다른']

# CountVectorizer 초기화
vectorizer = CountVectorizer(stop_words=korean_stop_words, max_features=50)  # 최대 50개의 키워드 추출

# 점수별로 리뷰를 그룹화
grouped_reviews = reviews.groupby('별점')['장점'].apply(lambda x: " ".join(x))

# 각 점수별 키워드 빈도 계산
score_keywords = {}
for score, text in grouped_reviews.items():
    # CountVectorizer로 텍스트 변환 및 키워드 추출
    X = vectorizer.fit_transform([text])
    keywords = vectorizer.get_feature_names_out()
    counts = X.toarray().flatten()
    
    # 키워드와 빈도를 딕셔너리로 저장
    score_keywords[score] = sorted(
        zip(keywords, counts), key=lambda x: x[1], reverse=True
    )

# 결과 출력
for score, keywords in score_keywords.items():
    print(f"점수 {score}의 주요 키워드:")
    for word, count in keywords:
        print(f"{word}: {count}")
    print("-" * 30)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score

data = pd.read_excel('데이터셋 (1).xlsx')
texts = data['장점']  # 텍스트 데이터 열 이름
labels = data['전/현직원']  # 레이블 데이터 열 이름

# 2. 데이터 전처리
# Vectorization with TF-IDF
tfidf = TfidfVectorizer(max_features=5000, stop_words='english')
X = tfidf.fit_transform(texts).toarray()

# 3. Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=42)

# 4. SVM 모델 학습
svm_model = SVC(kernel='linear')
svm_model.fit(X_train, y_train)

# 5. 모델 평가
y_pred = svm_model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
